```python
# from google.colab import files
# from IPython.display import Image
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# gambar_sakura = cv2_imshow.imread('C:\\Users\\ainin\\Downloads\\Sakura.jpg')
from google.colab.patches import cv2_imshow
# img_tomat = cv2.imread('tomat.jpg', cv2.IMREAD_UNCHANGED)
img = cv2.imread('Sakura.jpg', cv2.IMREAD_UNCHANGED)
# cv2_imshow(img)
# grayImg = cv2.cvtColor(np.array(img),cv2.COLOR_BGR2GRAY);
# cv2_imshow(grayImg)

color = ('b','g','r')
```

```python
# cv2_imshow('Original',gambar_tomat)
for i, col in enumerate(color):

    histr = cv2.calcHist([img],
                         [i], None,
                         [256],
                         [0, 256])

    plt.plot(histr, color = col)

    # Limit X - axis to 256
    plt.xlim([0, 256])
plt.show()
```

```python
height, width, _ = img.shape
```

```python
for i in range(0, height - 1):
    for j in range(0, width - 1):

        # Get the pixel value
        pixel = img[i, j]

        # Negate each channel by
        # subtracting it from 255

        # 1st index contains red pixel
        pixel[0] = 255 - pixel[0]

        # 2nd index contains green pixel
        pixel[1] = 255 - pixel[1]

        # 3rd index contains blue pixel
        pixel[2] = 255 - pixel[2]

        # Store new values in the pixel
        img[i, j] = pixel
```

```python

plt.imshow(img)
plt.show()

```

```python
for i, col in enumerate(color):

    histr = cv2.calcHist([img],
                         [i], None,
                         [256],
                         [0, 256])

    plt.plot(histr, color = col)
    plt.xlim([0, 256])

plt.show()
```
